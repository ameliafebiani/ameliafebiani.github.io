```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca file CSV
data = pd.read_csv('kebutuhan_per_bulan_2.csv')

# Membuat scatter plot
plt.scatter(data['kebutuhan'], data['tahun'])

# Memberikan judul dan label sumbu
plt.title('Scatter Plot Tahun dan Kebutuhan')
plt.xlabel('tahun')
plt.ylabel('kebutuhan')

# Menampilkan scatter plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Membaca data dari DataFrame
data = pd.DataFrame({
    'Propinsi': ['ACEH', 'SUMATERA UTARA', 'SUMATERA BARAT', 'RIAU', 'JAMBI', 'SULAWESI BARAT', 'MALUKU', 'MALUKU UTARA', 'PAPUA BARAT', 'PAPUA'],
    'Tahun': [2005, 2005, 2005, 2005, 2005, 2015, 2015, 2015, 2015, 2015],
    'Kebutuhan': [619876, 547255, 501315, 551498, 495242, 1981507, 2197450, 2333166, 2255113, 2171944]
})

# Membuat scatter plot
plt.scatter(data['Tahun'], data['Kebutuhan'])
plt.plot(data['Tahun'], data['Kebutuhan'], color='red', linestyle='--')
plt.xlabel('Tahun')
plt.ylabel('Kebutuhan')
plt.title('Scatter Plot Kebutuhan per Tahun')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca file CSV
data = pd.read_csv('profile_pelanggan.csv')

# Membuat scatter plot
plt.scatter(data['15572'], data['DKI Jakarta'])

# Memberikan judul dan label sumbu
plt.title('Scatter Plot 15572 dan DKI Jakarta')
plt.xlabel('15572')
plt.ylabel('DKI Jakarta')

# Menampilkan scatter plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Import data from CSV
data = pd.read_csv('profile_pelanggan.csv', header=None, names=['Kebutuhan', 'Propinsi'])

# Create scatter plot
plt.scatter(data['Propinsi'], data['Kebutuhan'])
plt.xlabel('Propinsi')
plt.ylabel('Kebutuhan')
plt.title('Scatter Plot Kebutuhan per Propinsi')
plt.xticks(rotation=90)
plt.show()
```
